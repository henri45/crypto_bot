```python
import pandas as pd
import numpy as np
```

## Read the historical data

```python
#Jan, March, April 2020

dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p')
data = pd.read_csv("Kraken_BTCEUR_1h.csv", skiprows = [0], parse_dates=['Date'], date_parser = dateparse)
data = data[(data['Date'] > "2019-01-01") & (data['Date'] < "2020-05-31")]
data = data[::-1]
data.reset_index(inplace = True, drop = True)
```

```python
week_placement = 300
balance_btc = 0
balance_euro = 3000
transaction_columns = ["Order", "Date", "btc_price", "btc_quantity", "threshold", "is_sold", "balance_eur", "balance_btc", "id_sold"]
transaction = pd.DataFrame(columns = transaction_columns)
transaction.loc[0, "balance_eur"] = balance_euro
transaction.loc[0, "balance_btc"] = balance_btc
profit = 1.15

for i in data.index[1:]:
    
    btc_price = data.loc[i,"Close"]
    
    date = data.loc[i,"Date"]
    
    #Buy the BTC
        
    if (data.loc[i,"Date"].weekday() == 6) & (data.loc[i,"Date"].hour == 7):

        btc_quantity = week_placement / btc_price

        threshold = btc_price * profit

        new_balance_eur = transaction.iloc[-1]["balance_eur"] - week_placement
        
        new_balance_btc = transaction.iloc[-1]["balance_btc"] + btc_quantity

        new_entry = ["Buy", date, btc_price, btc_quantity, threshold, False, new_balance_eur, new_balance_btc, np.nan]

        transaction = transaction.append(dict(zip(transaction_columns, new_entry)), ignore_index = True)
        
    #Sell the BTC
    
    to_sell = transaction[(transaction["threshold"] < btc_price) & (transaction["is_sold"] == False ) & (transaction["Order"] == "Buy" )]
    
    idx_to_sell = to_sell.index
    
    if len(idx_to_sell) > 0:
        
        for j in idx_to_sell:
            
            btc_quantity = to_sell.loc[j, "btc_quantity"]

            revenues = btc_quantity * btc_price * 0.98

            transaction.loc[j, "is_sold"] = True

            new_balance_eur = transaction.iloc[-1]['balance_eur'] + revenues
            
            new_balance_btc = transaction.iloc[-1]['balance_btc'] - btc_quantity

            threshold = to_sell.loc[j,"threshold"]

            new_entry = ["Sold", date, btc_price, btc_quantity, threshold, np.nan, new_balance_eur, new_balance_btc,  j]
                
            transaction = transaction.append(dict(zip(transaction_columns, new_entry)), ignore_index = True)
            
    if i % 100 == 0:
        print(i)
```
